```python
import pandas as pd
import numpy as np
import os
```

```python
dataset = pd.read_csv('train.csv')
indices = np.array(dataset.index)
np.random.shuffle(indices)
multiples = [k for k in range(1, len(indices)) if len(indices) % k == 0]
print(multiples)
k = 5
```

```python
k_folds = indices.reshape((k, len(indices)//k))
try:
    os.makedirs(f'./{k}_folds')
except:
    pass
for i, folds in enumerate(k_folds):
    dataset.iloc[folds].to_csv(f'./{k}_folds/{i}.csv',index=None)
```
