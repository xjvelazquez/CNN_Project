```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import os
%matplotlib inline
```

```python
def histogram_eq(input_im):
    hsv_im = cv2.cvtColor(input_im, cv2.COLOR_RGB2HSV)
    lookup = np.zeros((256))
    height, width = hsv_im[:,:,2].shape
    total = width*height
    cumulative = 0
    raw_hist = Counter(hsv_im[:,:,2].ravel())
    for i in range(256):
        cumulative += raw_hist[i]
        lookup[i] = cumulative / total
    for i in range(height):
        for j in range(width):
            hsv_im[i,j,2] = np.round(255*lookup[hsv_im[i,j,2]])
    return cv2.cvtColor(hsv_im,cv2.COLOR_HSV2RGB)
```

```python
tire = cv2.imread('tire_original.png')
plt.imshow(tire)
plt.figure()
plt.imshow(histogram_eq(tire))
```

```python
source = '/datasets/cs154-fa19-public/split_Ori_Face/test'
dest = './split_Ori_Face_equalized/test'
subjects = os.listdir(source)
for subject in subjects:
    print(subject)
    for picture in os.listdir(f'{source}/{subject}'):
        input_im = cv2.imread(f'{source}/{subject}/{picture}')
        try:
            os.makedirs(f'{dest}/{subject}/')
        except:
            pass
        try:
            if os.path.isfile(f'{dest}/{subject}/{picture}'):
                continue
            cv2.imwrite(f'{dest}/{subject}/{picture}',histogram_eq(input_im))
        except:
            print(f'error with: {subject}/{picture}')


        
```
